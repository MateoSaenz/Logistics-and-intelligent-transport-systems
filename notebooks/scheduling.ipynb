```python
import os

# Define the source and destination directories
source_dir = 'C:/Users/mateo/opl/Projet1/INSTANCES/INSTANCES'  # Replace with your source folder path
destination_dir = 'C:/Users/mateo/OneDrive/Documentos/IMT NORD EUROPE/AO 2/Logistics and sim/Datasets scheduling'  # Replace with your destination folder path

# Ensure the destination directory exists
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# Function to process each .txt file
def process_file(file_path):
    try:
        # Read the contents of the file
        with open(file_path, 'r') as file:
            lines = file.readlines()
        
        # Ensure the file has at least 3 lines
        if len(lines) < 3:
            print(f"Skipped {file_path}: insufficient data.")
            return None
        
        # Parse the first number as n and the second as m
        n = int(lines[0].strip())
        m = int(lines[1].strip())
        
        # Parse the third line as a list of integers
        pi = list(map(int, lines[2].strip().split()))
        
        # Convert to the desired format
        formatted_content = f"n = {n};\nm = {m};\nnpi = {pi};"
        return formatted_content
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

# Process all .txt files in the source directory
file_index = 1  # Start the file naming index from 1
for filename in os.listdir(source_dir):
    if filename.endswith('.txt'):
        # Get the full path of the file
        file_path = os.path.join(source_dir, filename)
        
        # Process the file to get the formatted content
        formatted_content = process_file(file_path)
        
        if formatted_content:
            # Create the new filename with the desired format
            new_filename = f"file_schedule_{file_index}.dat"
            new_file_path = os.path.join(destination_dir, new_filename)
            
            # Write the formatted content to the new file in the destination folder
            with open(new_file_path, 'w') as new_file:
                new_file.write(formatted_content)
            
            print(f"Processed {filename} -> {new_filename}")
            
            # Increment the file index
            file_index += 1

print("Processing complete.")
```

```python
import pandas as pd
import re

def process_files(processed_file, filename_file):
    # Step 1: Parse the processed_file
    processed_data = []
    with open(processed_file, 'r') as f:
        for line in f:
            match = re.match(r'Processed (\S+) -> (\S+)', line)
            if match:
                processed_data.append({
                    'Nom': match.group(1),
                    'Nom.dat': match.group(2),
                })

    # Step 2: Parse the filename_file
    filename_data = []
    with open(filename_file, 'r') as f:
        file_info = {}
        for line in f:
            if line.startswith("Filename."):
                if file_info:  # Save previous data
                    filename_data.append(file_info)
                file_info = {'FileName': int(line.split()[1])}
            elif line.startswith("UB:"):
                file_info['UB (Obj value)'] = float(line.split(":")[1].strip())
            elif line.startswith("LB:"):
                file_info['LB (Lower Bound)'] = float(line.split(":")[1].strip())
            elif line.startswith("Gap :"):
                file_info['Gap'] = float(line.split(":")[1].strip())
            elif line.startswith("Temps d'excution :"):
                file_info['Temps d\'excution'] = float(line.split(":")[1].strip())
        if file_info:  # Save the last set of data
            filename_data.append(file_info)

    # Step 3: Merge the data
    processed_df = pd.DataFrame(processed_data)
    filename_df = pd.DataFrame(filename_data)

    # Extract FileName from 'Nom.dat' to facilitate merging
    processed_df['FileName'] = processed_df['Nom.dat'].str.extract(r'_(\d+)\.dat').astype(int)

    # Merge on FileName
    final_df = pd.merge(processed_df, filename_df, on='FileName')

    # Step 4: Return the final DataFrame
    return final_df

# Example usage
processed_file = 'C:/Users/mateo/OneDrive/Documentos/IMT NORD EUROPE/AO 2/Logistics and sim/Processed documents.txt'  # Replace with the actual path to your file
#"C:/Users/mateo/OneDrive/Documentos/IMT NORD EUROPE/AO 2/Logistics and sim/Processed documents.txt"

filename_file = 'C:/Users/mateo/OneDrive/Documentos/IMT NORD EUROPE/AO 2/Logistics and sim/SAENZ_ORTIZ_Mateo_Lab_4/results_scheduling.txt'   # Replace with the actual path to your file
# "C:/Users/mateo/OneDrive/Documentos/IMT NORD EUROPE/AO 2/Logistics and sim/SAENZ_ORTIZ_Mateo_Lab_4/results_scheduling.txt"


result_df_documents = process_files(processed_file, filename_file)
result_df_documents
```

```python
import os
import pandas as pd
from docplex.mp.model import Model

# Set CPLEX solver path
os.environ['PATH'] += r';C:\\Program Files\\IBM\\ILOG\\CPLEX_Studio2211\\cplex\\bin\\x64_win64'

# Folder containing .txt files
folder_path = r'C:/Users/mateo/opl/Projet1/INSTANCES/testpy'

# Initialize a list to store results
results = []

# Function to solve the Parallel Machine Scheduling Problem
def solve_parallel_scheduling(m, n, processing_times, time_limit=15):
    # Create a CPLEX model
    mdl = Model(name="Parallel_Machine_Scheduling")
    
    # Set a time limit for the solver (in seconds)
    mdl.parameters.timelimit = time_limit

    # Decision variables: x[i][j] = 1 if task i is assigned to machine j
    x = mdl.binary_var_matrix(keys1=range(n), keys2=range(m), name="x")

    # Decision variable: Cmax (makespan)
    Cmax = mdl.continuous_var(name="Cmax")

    # Objective: Minimize Cmax
    mdl.minimize(Cmax)

    # Constraint: Each task must be assigned to exactly one machine
    for i in range(n):
        mdl.add_constraint(mdl.sum(x[i, j] for j in range(m)) == 1, ctname=f"task_{i}_assignment")

    # Constraint: Load on each machine cannot exceed Cmax
    for j in range(m):
        mdl.add_constraint(
            mdl.sum(processing_times[i] * x[i, j] for i in range(n)) <= Cmax,
            ctname=f"machine_{j}_load"
        )

    # Solve the model
    solution = mdl.solve(log_output=True)

    if solution:
        return solution.objective_value, [
            [int(x[i, j].solution_value) for j in range(m)] for i in range(n)
        ]
    else:
        return None, None

# Loop through each .txt file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.txt'):
        file_path = os.path.join(folder_path, file_name)
        
        # Log progress
        print(f"Processing file: {file_name}")

        # Read the file
        with open(file_path, 'r') as file:
            lines = file.readlines()
            m = int(lines[0].strip())  # Number of machines
            n = int(lines[1].strip())  # Number of tasks
            # Correctly parse processing times by splitting on whitespace
            processing_times = list(map(int, lines[2].strip().split()))

        # Solve the problem with a 15-second time limit
        optimal_makespan, task_assignment = solve_parallel_scheduling(m, n, processing_times, time_limit=15)

        # Store the results
        results.append({
            'File': file_name,
            'Machines': m,
            'Tasks': n,
            'Processing Times': processing_times,
            'Optimal Makespan': optimal_makespan,
            'Task Assignment': task_assignment
        })

# Create a DataFrame to display the results
results_df = pd.DataFrame(results)

# Save results to a CSV file
output_csv_path = os.path.join(folder_path, 'scheduling_results.csv')
results_df.to_csv(output_csv_path, index=False)

# Print the results
results_df
```

```python
import os
import pandas as pd
from docplex.mp.model import Model

os.environ['PATH'] += r';C:\\Program Files\\IBM\\ILOG\\CPLEX_Studio2211\\cplex\\bin\\x64_win64'

folder_path = r'C:/Users/mateo/opl/Projet1/INSTANCES/testpy'

results = []

def solve_parallel_scheduling(m, n, processing_times, time_limit=15):
    
    mdl = Model(name="Parallel_Machine_Scheduling")
    
   
    mdl.parameters.timelimit = time_limit

    # Decision variables: x[i][j] = 1 if task i is assigned to machine j
    x = mdl.binary_var_matrix(keys1=range(n), keys2=range(m), name="x")

    # Decision variable: Cmax 
    Cmax = mdl.continuous_var(name="Cmax")

    # Objective: Minimize Cmax
    mdl.minimize(Cmax)

    # Constraint: Each task must be assigned to exactly one machine
    for i in range(n):
        mdl.add_constraint(mdl.sum(x[i, j] for j in range(m)) == 1, ctname=f"task_{i}_assignment")

    # Constraint: Load on each machine cannot exceed Cmax
    for j in range(m):
        mdl.add_constraint(
            mdl.sum(processing_times[i] * x[i, j] for i in range(n)) <= Cmax,
            ctname=f"machine_{j}_load"
        )

    # Solve the model
    solution = mdl.solve(log_output=True)

    # Collect metrics
    if solution:
        return {
            'Optimal Makespan': solution.objective_value,
            'Upper Bound (UB)': mdl.objective_value,
            'Lower Bound (LB)': mdl.solve_details.best_bound,
            'Gap': mdl.solve_details.mip_relative_gap,
            'Execution Time': mdl.solve_details.time,
            'Task Assignment': [
                [int(x[i, j].solution_value) for j in range(m)] for i in range(n)
            ]
        }
    else:
        return {
            'Optimal Makespan': None,
            'Upper Bound (UB)': None,
            'Lower Bound (LB)': None,
            'Gap': None,
            'Execution Time': None,
            'Task Assignment': None
        }

# Loop through each .txt file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.txt'):
        file_path = os.path.join(folder_path, file_name)
        
        # Log progress
        print(f"Processing file: {file_name}")

        # Read the file
        with open(file_path, 'r') as file:
            lines = file.readlines()
            m = int(lines[0].strip())  # Number of machines
            n = int(lines[1].strip())  # Number of tasks
            processing_times = list(map(int, lines[2].strip().split()))

        # Solve the problem with a 15-second time limit
        metrics = solve_parallel_scheduling(m, n, processing_times, time_limit=15)

        # Store the results
        results.append({
            'File': file_name,
            'Machines': m,
            'Tasks': n,
            'Processing Times': processing_times,
            **metrics
        })

results_df = pd.DataFrame(results)

output_csv_path = os.path.join(folder_path, 'scheduling_results.csv')
results_df.to_csv(output_csv_path, index=False)

results_df
```

```python
def compare_dataframes(df1, df2):

  
    results = {}
    
    # Check if both DataFrames have the same number of rows
    if len(df1) != len(df2):
        raise ValueError("The DataFrames must have the same number of rows to compare.")
    
    # Compare UB row-by-row
    ub_differences = []
    for i in range(len(df1)):
        if df1.loc[i, 'UB'] != df2.loc[i, 'UB']:
            ub_differences.append({
                'Filename': df1.loc[i, 'Filename'],
                'UB_df1': df1.loc[i, 'UB'],
                'UB_df2': df2.loc[i, 'UB']
            })
    
    results['UB_Differences'] = ub_differences
    
    # Calculate average (mean) values for Gap and Execution Time
    gap_mean_df1 = df1['Gap'].mean()
    gap_mean_df2 = df2['Gap'].mean()
    exec_time_mean_df1 = df1['Execution Time'].mean()
    exec_time_mean_df2 = df2['Execution Time'].mean()
    
    results['Gap_Mean_Comparison'] = {
        'Gap_Mean_df1': gap_mean_df1,
        'Gap_Mean_df2': gap_mean_df2,
        'Difference': gap_mean_df2 - gap_mean_df1
    }
    
    results['Execution_Time_Mean_Comparison'] = {
        'Execution_Time_Mean_df1': exec_time_mean_df1,
        'Execution_Time_Mean_df2': exec_time_mean_df2,
        'Difference': exec_time_mean_df2 - exec_time_mean_df1
    }
    
    return results

comparison_results = compare_dataframes(df_iloc, results_df)

for key, value in comparison_results.items():
    print(f"{key}: {value}")
```
